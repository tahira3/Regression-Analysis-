<h4>Multiple Linear Regression</h4>

```python
lm = LinearRegression()
lm
```

```python
Z = df[['family member', 'cooking frequency', 'usage score', 'yearly occasions (disscount factor)']]
lm.fit(Z, df['Gas tank longivity( days)'])
```

```python
lm.intercept_
```

```python
lm.coef_
```

<h4>R-square value for multiple linear regression</h4>

```python
print('The R-square value is: ', lm.score(Z, df['Gas tank longivity( days)']))
```

we have got r-square value of 0.98 that indicates that our model is 98% accurate and we can use the model to predict gas tank longitivity with respect to the variable family member,cooking frequency,usage score and yearly occasions.<br></br>
Here's our model,<br></br>   Y=58.35+(-3.72*a)+(-6.78*b)+(0.67*c)+(-0.32*d)<br></br>
a=family member<br></br>
b=cooking frequency<br></br>
c=usage score<br></br>
d=yearly occasions (disscount factor)<br></br>

```python
Y_hat = lm.predict(Z)
```

```python
plt.figure(figsize=(width, height))


ax1 = sns.distplot(df['Gas tank longivity( days)'], hist=False, color="g", label="Actual Value")
sns.distplot(Y_hat, hist=False, color="b", label="Predicted Values" , ax=ax1)



plt.title('Actual vs Pridicted Values for Gas tank longivity( days)')
# plt.xlabel('Factors')
# plt.ylabel('Gas tank longivity( days)')
plt.legend()

plt.show()
```
