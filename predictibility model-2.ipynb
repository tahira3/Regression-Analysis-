<h4>Multiple Linear Regression</h4>

```python
Z = df[['Family member', 'Cooking frequency', 'Usage score', 'Yearly occasions (disscount factor)','Birthday ( discount factor )']]
lm.fit(Z, df['Gas tank longivity( days)'])
```

```python
lm.intercept_
```

```python
lm.coef_
```

<h4>R-square value for multiple linear regression</h4>

```python
print('The R-square value is: ', lm.score(Z, df['Gas tank longivity( days)']))
```

we have got r-square value of 0.97 that indicates that our model is 97% accurate and we can use the model to predict gas tank longitivity with respect to the variable family member,cooking frequency,usage score and yearly occasions.<br></br>
Here's our model,<br></br>   Y=65.5+(-3.31*a)+(-9.05*b)+(1.14*c)+(-1.08*d)+(-3.31*e)<br></br>
a=Family member<br></br>
b=Cooking frequency<br></br>
c=Usage score<br></br>
d=Yearly occasions (disscount factor)<br></br>
e=Birthday ( discount factor )

```python
Y_hat = lm.predict(Z)
```

```python
plt.figure(figsize=(width, height))


ax1 = sns.distplot(df['Gas tank longivity( days)'], hist=False, color="g", label="Actual Value")
sns.distplot(Y_hat, hist=False, color="b", label="Pridicted Values" , ax=ax1)



plt.title('Actual vs Pridicted Values for Gas tank longivity( days)')
# plt.xlabel('Factors')
# plt.ylabel('Gas tank longivity( days)')
plt.legend()

plt.show()
plt.close()
```
